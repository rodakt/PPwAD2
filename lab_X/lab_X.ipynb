Napisz moduł `moving_average.py`  z klasami `AverageSpeed` i `MovingAverage`. 

Instancje klasy `MovingAverage` mają przechowywać i aktualizować średnią ruchomą na podstawie dodawanych wartości. Do interfejsu klasy należą atrybuty:
* `total` -- wartość liczbowa określająca sumę dodanych wartości;
* `count` -- wartość liczbowa określająca liczbę dodanych wartości;

i metody:

* `__init__(self)` -- inicjalizuje atrybuty `total` i `count` na zero;

* `distance` -- wartość liczbowa określająca przebytą drogę;
* `time` -- wartość liczbowa określająca czas zużyty na przebycie drogi;
* `average_speed` --  prędkość średnia punktu;

i metody:

* `__init__(self)` -- inicjalizuje atrybuty `distance` i `time` na zero;
* `add_section(self, distance, time)` -- aktualizuje atrybuty na podstawie nowych danych, oblicza prędkość średnią i przypisuje ją do atrybutu `average_speed`.
