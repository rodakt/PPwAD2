#### Wskazówka 

W metodzie `__init__` klasy `Intervals` zaimplementuj algorytm scalania przedziałów. W tym celu zdefiniuj funkcję `merge_intervals(intervals)`, która przyjmuje listę obiektów klasy `OpenInterval` i zwraca listę obiektów klasy `OpenInterval` będących scalonymi przedziałami z listy `intervals`. Oto rekurencyjny algorytm scalania w pseudokodzie:

```
merge_intervals(intervals):
    for A, B in kombinacje(intervals, 2):
        if A & B:
            intervals.remove(A)
            intervals.remove(B)
            intervals.append(A | B)
            return merge_intervals(intervals)
    return intervals
```

## Zadanie XI.5

Zaprojektuj klasę `ZZ` implementującą arytmetykę reszt modulo $m\geqslant 2$ na zbiorze reszt $\{0,1,\dots,m-1\}$. Arytmetyka ta wygląda następująco: jeśli $a,b\in\{0,1,\ldots,m-1\}$, to 
* $a\oplus b$ jest resztą z dzielenia $a+b$ przez m;
* $a\odot b$ jest resztą z dzielenia $a\cdot b$ przez m.

Na przykład, jeśli $m=7$, to zbiór reszt jest równy $\{0,1,2,3,4,5,6\}$. Wtedy dla $a=5$, $b=2$ mamy
$$a\oplus b=0, \quad a\odot b =3$$ 

Przez $-a$ oznaczamy taką resztę $b$, że $a\oplus b=0$. Podobnie, przez $1/a$ oznaczamy taką resztę $b$, że $a\odot b=1$.
Reszta $-a$ zawsze istnieje, reszta $1/a$ może nie istnieć.

Odejmowanie $\ominus$ definiujemy wzorem

$$a\ominus b:=a\oplus(-b).$$

Ponadto, jeśli $k$ jest liczbą całkowitą, to  $k\cdot a$ oznacza:
  * $a\oplus a \oplus\cdots\oplus a$ -- suma $k$ składników, gdy $k\geqslant 0$;
  * $-(a\oplus a \oplus\cdots\oplus a)$ -- element przeciwny do sumy $-k$ składników, gdy $k<0$.

Przykładowo, przy podanych wyżej oznaczeniach

$$\quad a\oplus 2\cdot b=2,\quad 10\cdot a=1,\quad -4\cdot b=6.$$

W iloczynie liczby całkowitej przez resztę liczba całkowita jest zawsze pierwszym czynnikiem.

Potęgowanie definiujemy następująco:
* $a^0=1$ dla $a\neq 0$,
* $a^k=a\odot a\odot\cdots\odot a$ -- iloczyn $k$ składników, gdy $k>0$,

Naiwna implementacja potęgowania wymaga iteracyjnego mnożenia reszt przez siebie. W przypadku dużych wykładników jest to bardzo czasochłonne. Można jednak zastosować [algorytm szybkiego potęgowania](https://en.wikipedia.org/wiki/Exponentiation_by_squaring), który pozwala na obliczenie $a^k$ w czasie $O(\log k)$. Oto przykład ręcznego obliczania $3^{101}$ w arytmetyce reszt modulo 7:

$$
3^{101}\equiv 3^{2\cdot 50 + 1}\equiv (3^2)^{50}\cdot 3\equiv 9^{50}\cdot 3\equiv 2^{50}\cdot 3\\
\equiv 4^{25}\cdot 3\equiv 4^{2\cdot 12 + 1}\cdot 3\equiv 16^{12}\cdot 4\cdot 3\equiv 2^{12}\cdot 12\\
\equiv 4^6\cdot 5\equiv 16^3\cdot 5\equiv 2^3\cdot 5\equiv 5
$$
